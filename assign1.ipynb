```python
import csv
import requests
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import string

# Download NLTK resources
import nltk
nltk.download('punkt')
nltk.download('stopwords')

def download_content_from_api(title):
    endpoint = "https://en.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "format": "json",
        "titles": title,
        "prop": "extracts",
        "exintro": True,
    }

    response = requests.get(endpoint, params=params)
    data = response.json()

    page = next(iter(data["query"]["pages"].values()))
    if "extract" in page:
        return page["extract"]
    else:
        print(f"The page with the title '{title}' does not exist on Wikipedia.")
        return None
    

document1_title = "Artificial_intelligence"

document1_content = download_content_from_api(document1_title)

```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
import unicodedata
from bs4 import BeautifulSoup

nltk.download('wordnet')

def preprocess_and_tokenize(text):
    # Remove HTML tags
    soup = BeautifulSoup(text, "html.parser")
    text = soup.get_text(separator=" ")

    # Normalize text (convert to lowercase and remove accents)
    text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8').lower()

    # Remove punctuation
    text = text.translate(str.maketrans("", "", string.punctuation))
    text = text.replace('\n', '')

    # Tokenize and remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return tokens , text



document1_tokens , text = preprocess_and_tokenize(document1_content)

text

```
